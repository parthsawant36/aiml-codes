```python

import numpy as np


y_true = [0, 1, 0, 1, 0, 1, 1, 0, 1, 0]
y_pred = [0, 1, 0, 0, 0, 1, 1, 0, 1, 1]
```

```python
tp = 0
tn = 0
fp = 0
fn = 0
```

```python
for true, pred in zip(y_true, y_pred):
    if true == 1 and pred == 1:
        tp += 1
    elif true == 0 and pred == 0:
        tn += 1
    elif true == 0 and pred == 1:
        fp += 1
    elif true == 1 and pred == 0:
        fn += 1
```

```python
confusion_matrix = [[tn, fp], [fn, tp]]
```

```python
for row in confusion_matrix :
    print(row)
```

```python
#Evaluation Metric
```

```python
#accuracy
acc1 = tp+tn
acc2 = tp+tn+fp+tn

accuracy = acc1/acc2
```

```python
#recall
re1 = tp 
re2 = tp+fn
recall = re1/re2
```

```python
#precision
pre1=tp
pre2=tp+fp
precision = pre1 / pre2
```

```python
#F1-Score
sc1 = 2*precision*recall
sc2 = precision+recall
f1 = sc1/sc2
```

```python
print("Evaluation Metrics")
print(f"\nAccuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```

```python
#Error Metrics
```

```python
sub_error = 0
sq_error = 0
n = len(y_true)
```

```python
for i in range(n) :
    error = y_true[i] - y_pred[i]

    #MAE
    sub_error = sub_error + abs(error)

    #MSE
    sq_error = sq_error + np.power(error,2)


mae = sub_error / n

mse = sq_error / n

#rmse
rmse = np.sqrt(mse) 

print("Error Metrics")
print(f"\nMean Absolute Error: {mae}")
print(f"Mean Sqaured Error: {mse}")
print(f"Root Mean Sqaured Error: {rmse}")
```
